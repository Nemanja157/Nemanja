***PODACI:***

Ovaj kod se odnosi na pripremu podataka za obuku i evaluaciju modela dubokog učenja za detekciju godina, pola i emocija na slikama. Korišćen je ImageDataGenerator iz TensorFlow-ove biblioteke kako bi se efikasno normalizovali pikseli slika, skalirali u opseg od 0 do 1, i izvršila eventualna augmentacija podataka.

Zatim su generisani podaci za obuku i validaciju pomoću funkcije flow_from_directory, koja automatski čita slike iz odgovarajućih podfoldera prema zadatoj putanji data_directory. Podaci su podeljeni u dva skupa, jedan za obuku i drugi za validaciju, pri čemu 80% podataka ide u skup za obuku, a 20% u skup za validaciju.

Ovaj pristup omogućava efikasno učenje modela na slikama koje predstavljaju različite klase godina, pola i emocija, dok se istovremeno osigurava da model bude sposoban za dobro generalizovanje na nepoznate podatke. Spremnost podataka je ključna za uspešno treniranje modela dubokog učenja.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

def preprocess_data(data_dir, img_size=(256, 256), batch_size=32):
    # Korišćenje ImageDataGenerator za normalizaciju i augmentaciju podataka
    datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

    # Generisanje podataka za obuku
    train_generator = datagen.flow_from_directory(
        data_dir,
        target_size=img_size,
        batch_size=batch_size,
        class_mode='categorical',
        subset='training'
    )

    # Generisanje podataka za validaciju
    validation_generator = datagen.flow_from_directory(
        data_dir,
        target_size=img_size,
        batch_size=batch_size,
        class_mode='categorical',
        subset='validation'
    )

    return train_generator, validation_generator
```

**MODEL:**

Ovaj kod definiše arhitekturu modela za detekciju godina, pola i emocija na osnovu slika. Model je projektovan da automatski nauči relevantne karakteristike iz podataka koristeći konvolutivne neuronske mreže (CNN). Arhitektura modela uključuje tri konvolutivna sloja (Conv2D) koji služe za ekstrakciju različitih nivoa apstrakcije iz ulaznih slika. Slojevi grupisanja (MaxPooling2D) se koriste kako bi se smanjila prostorna dimenzionalnost i ubrzao proces obučavanja.

Nakon konvolutivnih slojeva, dodat je sloj globalnog prosečnog objedinjavanja kako bi se pojednostavio izlaz i dobile značajke relevantne za celu sliku. Zatim sledi gusti sloj sa 1024 neurona i funkcijom aktivacije ReLU, koji služi za dalju apstrakciju i obogaćivanje informacija. Konačno, dodat je gusti sloj sa softmax aktivacijom, koji vrši klasifikaciju u više klasa, odgovarajući različitim kategorijama godina, pola i emocija.

Model koristi Adam optimizator za prilagodljivu stopu učenja, a funkciju gubitka postavlja na 'categorical_crossentropy', što je standardni izbor za klasifikacione zadatke sa više klasa. Metrika tačnosti (accuracy) prati performanse modela tokom obučavanja. Ovaj model može biti prilagođen i dalje usavršen u skladu sa specifičnostima i zahtevima projekta.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.optimizers import Adam

def build_custom_model(input_shape, num_classes):
    # Kreiranje modela
    model = Sequential()

    # Dodavanje slojeva
    model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(GlobalAveragePooling2D())
    model.add(Dense(1024, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))

    # Kompilacija modela
    model.compile(optimizer=Adam(lr=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

    return model
```

***DVE VRSTE KLASA:***


Ovaj deo koda služi za analizu distribucije klasa u skupu podataka, što je važno za razumevanje ravnoteže ili neravnoteže podataka. Ovaj proces ima nekoliko ključnih koraka:

**Izračunavanje broja klasa**: Korišćenjem np.unique, dobijamo niz koji sadrži broj slika za svaku klasu u trening skupu podataka. Ovo nam daje uvid u to koliko je svaka klasa zastupljena u skupu podataka.

**Prikazivanje bar grafikona**: Nakon što smo dobili informacije o broju slika po klasama, koristimo matplotlib.pyplot da vizualno prikažemo distribuciju klasa pomoću bar grafikona. Na x-osi su prikazane klase, dok je na y-osi broj slika. Ovo nam pomaže da brzo identifikujemo eventualnu neravnotežu u skupu podataka, gde neke klase mogu biti nedovoljno ili previše zastupljene u odnosu na druge.

Ovaj korak je ključan kako bismo pravilno balansirali skup podataka i izbegli pristrasnost u modelu prema određenim klasama.

```python
import numpy as np
import matplotlib.pyplot as plt

def class_distribution(data_generator):
    # Dobijanje broja slika po klasama iz generatora podataka
    class_counts = np.unique(data_generator.classes, return_counts=True)[1]
    class_labels = list(data_generator.class_indices.keys())

    # Prikazivanje bar grafikona
    plt.figure(figsize=(10, 6))
    plt.bar(class_labels, class_counts)
    plt.title('Distribucija klasa')
    plt.xlabel('Klase')
    plt.ylabel('Broj slika')
    plt.show()

# Pozivanje funkcije za trening generator
class_distribution(train_generator)
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

# Putanja do direktorijuma sa podacima
data_dir = "/content/drive/MyDrive/data"

# Dimenzije slika
img_size = (224, 224)
batch_size = 32

# Kreiranje generatora podataka sa normalizacijom i podelom na obuku i validaciju
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

# Učitavanje podataka za obuku
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

# Učitavanje podataka za validaciju
validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Kreiranje modela
base_model = ResNet50(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
base_model.trainable = False  # Zamrzavanje težina postojećeg modela

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dense(2, activation='softmax')  # Broj klasa
])

# Kompilacija modela
model.compile(optimizer=Adam(lr=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

# Treniranje modela
epochs = 8
history = model.fit(train_generator, epochs=epochs, validation_data=validation_generator)
```

***EVALUACIJA MODELA:***

Nakon izvršavanja ovog dela koda, dobićemo informacije o gubitku i tačnosti modela na test podacima. Ove metrike su važne za procenu performansi modela i identifikaciju potencijalnih oblasti za poboljšanja.







```python
# Evaluacija modela na test podacima
evaluation = model.evaluate(validation_generator)

# Ispis rezultata evaluacije
print("Loss: {:.2f}".format(evaluation[0]))
print("Accuracy: {:.2f}%".format(evaluation[1] * 100))
```
